```python
hiliary_data = bnames[(bnames['name']=='Hillary') & (bnames['sex']=='girl')]
sns.lineplot(x='year', y='percent', data=hiliary_data)
```

```python
## 4. Plot the popularity of 'Emma' (Female) and 'Ryan' (Male) by year on the same graph

emma_ryan_data = bnames[((bnames['name']=='?') & (bnames['sex']=='?')) | ((bnames['?']=='?') & (bnames['?']=='?'))]
sns.lineplot(x='?', y='?', data=emma_ryan_data, hue='?')
```
