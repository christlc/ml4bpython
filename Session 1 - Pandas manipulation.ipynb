# 5 Basic Data Manipulation
1. filter by rows
2. select columns
3. sort
4. add new columns
5. group by 

```python
# 1 Filter by row

euro[euro['Stamina'] < 20 ]
```

```python
# 2 Select Columns
euro[['Name', 'League', 'Stamina']]
```

```python
# Combine both Method
smaller_euro = euro[['Name', 'League', 'Position', 'Stamina']]
smaller_euro[smaller_euro['Stamina'] < 20]
```

```python
# more concise
euro[['Name', 'League', 'Position', 'Stamina']][euro['Stamina']<20]
```

```python
# 3 Sort
smaller_euro.sort_values('Stamina')
```

```python
smaller_euro.sort_values(['Stamina', 'Position'])
```

```python
?euro.sort_values
```

```python
smaller_euro.sort_values('Stamina', ascending=False)
```

```python
# 4 Add new Columns
euro['HeightM'] = euro['Height']/100
euro[['Name','HeightM', 'Height']]
```

```python
# Add BMI
euro["BMI"] = euro["Weight"]/(euro['HeightM'])**2
euro[["BMI"]]
```

```python
euro.sort_values('BMI', ascending=False)[['Name','BMI', 'Position']]
```

```python
sns.scatterplot('BMI', 'Stamina', data=euro, hue='Position')
```

```python
# 5 Making Summary 
```

```python
euro['BMI'].min()
```

```python
euro[['BMI', 'Stamina']].agg(['max', 'min', 'mean', 'median'])
```

```python
euro.groupby('Position')[['Stamina']].agg(['mean', 'std'])
```

```python
# Remember the outlier!
euro[euro['Stamina']>10].groupby('Position')[['Stamina']].agg(['mean', 'std'])
```

```python
euro.groupby('Position').agg('count')
```

```python
euro.groupby('Position').agg({'Name':'count'})
```

```python
# BONUS: can do complex query with group by and apply
euro[['Name', 'Position', 'Stamina']].groupby('Position').apply(lambda x: x.nlargest(3, ['Stamina']))
```
