```python
import pandas as pd
import seaborn as sns
%pylab inline
```

```python
bnames = pd.read_csv("bnames.csv")
```

```python
bnames.head()
```

```python
euro = pd.read_csv("euro.csv")
```

```python
euro.head()
```

```python
# Scatterplot
sns.scatterplot(x='Header', y='Jump', data=euro, alpha= 0.2)
```

```python
# Hexplots
sns.jointplot(x=euro['Header'], y=euro['Jump'], kind="hex", color="k")
```

```python
# Histograms
sns.distplot(bnames[(bnames['sex']=='boy') & (bnames['year']==2000)]['percent'], kde=False, rug=False)
```

```python
sns.distplot(euro['Stamina'], kde=False, rug=False)
```

```python
# Density Plot
```

```python
sns.distplot(euro['Stamina'], kde=True, rug=False)
```

```python
# Pair Plot
sns.pairplot(euro[['Stamina', 'Jump', 'Header', 'Mentality', 'Teamwork']],  plot_kws={'alpha': 0.2})
```

```python
# Regression Plot
df = euro[(euro['Position']=='Forward')]
sns.regplot(x='Header',y='Jump', data=df)
```
