```python
# Task 2 - Create a blackbox model for prediction
```

```python
# encode the data using pd.get_dummies
daily_encoded = pd.get_dummies(daily[['n','days', 'wday']],drop_first=True)
```

```python
y = daily_encoded['n']
X = daily_encoded.drop(['n'],axis=1)
```

```python
# Try fitting a linear regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
cross_val_score(reg, X, y, cv=5, scoring='neg_mean_squared_error').mean()
```
